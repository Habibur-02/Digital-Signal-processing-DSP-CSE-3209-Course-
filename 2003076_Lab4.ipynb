# ***Task 1: Take a sinusoid of a fundamental frequency and some of its harmonics. Show the generated periodic signal using the components.***

```python
import numpy as np
import matplotlib.pyplot as plt

# Time axis
t = np.linspace(0, 1, 1000)  # 1 second, 1000 points

# Fundamental frequency
f0 = 5  # 5 Hz

# Number of harmonics
k = 5  # total harmonics including fundamental

# Initialize total signal
x_total = np.zeros_like(t)

# ----------------- Plot Fundamental and Harmonics -----------------
plt.figure(figsize=(12, 8))

for i in range(1, k+1):
    amplitude = 1 / i  # amplitude reduces with harmonic number
    harmonic = amplitude * np.sin(2 * np.pi * i * f0 * t)
    x_total += harmonic  # Add to total combined signal

    # Plot each harmonic
    plt.plot(t, harmonic, label=f'{i}x fundamental ({i*f0} Hz)', alpha=0.7)

plt.title(f'Fundamental and {k-1} Harmonics')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()

# ----------------- Plot Combined Signal -----------------

plt.figure(figsize=(10, 5))
plt.plot(t, x_total, color='red')
plt.title(f'Combined Signal: Fundamental + {k-1} Harmonics')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

# ***Task 2:You are given the Fourier coefficients of a signal: ***
$C_k = \begin{cases}
\frac{Aa}{T}, & k=0 \\
\frac{Aa}{T} \, \text{sinc}(\pi k f a), & \text{otherwise}
\end{cases}
$

i. Plot the coefficients in the frequency domain. \
ii. Generate the power density spectrum. \
iii. Generate the original signal using the synthesis equation.

```python
# Parameters
A = 2         # Amplitude
T = 1         # Time period
a = 0.2       # Duration of non-zero signal
Fa = a / T    # Ratio

# Harmonics range
N = 50
k = np.arange(-N, N + 1)

# Compute CTFS coefficients
C_k = (A * a / T) * np.sinc(k * Fa)  # np.sinc(x) = sin(pi*x)/(pi*x)

# Plot |C_k|
plt.plot(k, C_k)
plt.title("CTFS Coefficients |Cₖ|")
plt.xlabel("k")
plt.ylabel("Cₖ")
plt.grid(True)
plt.show()
```

```python
# Power Spectrum
P_avg = np.abs(C_k) ** 2

# Plot P_avg
plt.plot(k, P_avg)
plt.title("Power Density Spectrum (|Cₖ|²)")
plt.xlabel("k")
plt.ylabel("P_avg")
plt.grid(True)
plt.show()
```
