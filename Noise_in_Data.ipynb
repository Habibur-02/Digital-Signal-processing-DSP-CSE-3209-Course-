```python
import numpy as np
import matplotlib.pyplot as plt
x=[1,2,3]
nx=np.arange(0,len(x),1)
plt.subplot(3,1,1)
plt.stem(nx,x)
plt.show()

h=[4,5,6]
nh=np.arange(0,len(h),1)
plt.subplot(3,1,2)
plt.stem(nh,h)
plt.show()



y=np.convolve(x,h)
print(y)
ny=np.arange(0,len(y),1)
plt.subplot(3,1,3)
plt.stem(ny,y)
plt.show()
```

```python
import numpy as np

signal1 = np.array([1, 2, 3])
signal2 = np.array([4,5,6])

convolution_full =np.convolve(signal1, signal2, mode='full')
convolution_valid = np.convolve(signal1, signal2, mode='valid')
convolution_same = np.convolve(signal1, signal2, mode='same')

print("Signal 1:", signal1)
print("Signal 2:", signal2)
print("Full Mode Convolution:", convolution_full)
print("Valid Mode Convolution:", convolution_valid)
print("Same Mode Convolution:", convolution_same)
```

```python
Srate = 256 # Hz
t = np.arange(0,3,1/Srate)
pnts = len(t)
print(pnts)
```

```python

#Create a noiseless signal...
x = np.cos(2*np.pi*1*t)
plt.plot(t,x)

```

```python
noise = np.random.randn(pnts)
plt.plot(t,noise)
print(noise)
```

```python
NoisySignal = x+noise

plt.plot(t,noise)
```

```python
filter = 2*np.ones(50)/50.0
print(filter)
```

```python
#Filtering/Noise-Cancellation by Convolution
filteredSignal = np.convolve(NoisySignal, filter)
```

```python
plt.figure(figsize=(12,12))
#style.use('dark_background')

plt.plot(filteredSignal, label='FIltered Signal')
plt.legend(fontsize=30)

plt.show()
```

```python
filteredSignal2 = np.convolve(filteredSignal, filter)
```

```python
plt.figure(figsize=(12,12))

plt.stem(filteredSignal2, label='FIltered Signal')
plt.legend(fontsize=30)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(0)
signal1 = np.sin(np.linspace(0, 2 * np.pi, 100))
signal2 = np.sin(np.linspace(0, 2 * np.pi, 100) + np.pi / 4)

signal1 += np.random.normal(0, 0.1, 100)
signal2 += np.random.normal(0, 0.1, 100)


correlation = np.correlate(signal1, signal2, mode='full')


lags = np.arange(-len(signal1) + 1, len(signal1))

plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(signal1, label='Signal 1')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(signal2, label='Signal 2', color='orange')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(lags, correlation, label='Correlation', color='green')
plt.xlabel('Lag')
plt.ylabel('Correlation')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0, 100)
signal1 = np.sin(0.1 * np.pi * n)
signal2 = np.sin(0.1 * np.pi * (n - 5))


signal1 += np.random.normal(0, 0.1, 100)
signal2 += np.random.normal(0, 0.1, 100)


correlation = np.correlate(signal1, signal2, mode='full')

lags = np.arange(-len(signal1) + 1, len(signal1))

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.stem(n, signal1)
plt.title('Signal 1')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.stem(n, signal2)
plt.title('Signal 2')
plt.xlabel('n')
plt.ylabel('Amplitude')

# Plot the correlation
plt.subplot(3, 1, 3)
plt.stem(lags, correlation)
plt.title('Cross-Correlation of Signal 1 and Signal 2')
plt.xlabel('Lag')
plt.ylabel('Correlation')

plt.tight_layout()
plt.show()
```

```python
NewSig = np.zeros(100)
NewSig[15:35]=1
NewSig[65:85]=1
plt.figure(figsize = (20,8))


plt.xlim(0,100)
plt.plot(NewSig,'o-', linewidth = 3, label = 'Signal with edges')
plt.legend(fontsize = 20)
plt.show()

# Kernel for edge detection

kernel = np.array([0.5,0.3,0.1,0,-0.1,-0.3,-0.5])
kernel

```

```python
# Plotting the kernel

#plt.figure(figsize = (20,8))
#plt.plot(kernel,'o-', linewidth=3, label = 'Kernel for edge detection')
#plt.legend(fontsize = 20)
#plt.show()
```

```python

filt_sig = np.convolve(NewSig,kernel, mode = 'same')
```

```python
# Plotting Filtered Signal

plt.figure(figsize = (20,8))
plt.plot(filt_sig,linewidth =3, label = 'Filtered Signal')
plt.legend(fontsize = 20)
plt.show()
```

```python
# Plotting original signal and filtered signal

plt.figure(figsize = (24,12))
plt.plot(NewSig,'o-', linewidth =3, label = 'Original Signal')
plt.plot(filt_sig,'r-',linewidth =3, label = 'Filtered Signal')
plt.legend(fontsize = 20)
plt.title('Edge Detection', fontsize = 30)
plt.show()
```
